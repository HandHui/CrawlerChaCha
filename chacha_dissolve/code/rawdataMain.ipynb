```python
length = len(policy_lst)
print(length)

clst = []
for i in range(length):
#     try:
    print(i)
    tdict = {}
    tdict['idx'] = i

    plist = PageList(policy_lst[i])
    plist.handle_content()

    alst = plist.content
    url = plist.content[1]

    dp = DetailPage(url)
    dp.handle_detail()

    nxurl = dp.next_url


    ks = ['title','url','area','agent','time','timespan']

    title = alst[0]
    tdict['title'] = title   ### 标题
    print(title)
#     print(title)

    tdict['url'] = alst[1]  ### url

    tt = re.split('[:：]',alst[2])  ## 适用地区
    if len(tt) < 2:
        tdict['area'] = ''
    else:
        tdict['area'] = tt[1].strip()


    tt = re.split('[:：]',alst[3])  ##有效期限 （起始与终止）
    if len(tt) < 2:
        tdict['timestart'] = "0000-00-00"
        tdict['timeend'] = "0000-00-00"
    else:
        tspan = tt[1]
        tspan = tspan.split('至')
        if len(tspan) == 2:
            tdict['timestart'] = tspan[0]
            tdict['timeend'] = tspan[1]
        else:      
            tdict['timestart'] = tt[1][:-3]
            tdict['timeend'] = "0000-00-00"


    sup_ind = dp.sup_ind  ## 适用行业
    tdict['sup_ind'] = sup_ind

    money = dp.money
    tdict['money'] = money

    dept = dp.source ## 对应责任部门
    tdict['dept'] = dept

    source = dp.nxtitle  ##3项目来源
    tdict['source'] = source

    mbody = dp.mbody
    if "申报条件" in mbody and "申报材料" in mbody:
        cond = S_search('申报条件(.*?)申报材料',mbody,re.S)
        tdict['condition'] = cond

        material = S_search('申报材料(.*)',mbody,re.S)
        tdict['material'] = material
    else:
        cond = dp.mbody
        tdict['condition'] = cond
        if nxurl.endswith('='):
            tdict['material'] = ""  ## https://www.chacha.top/sup_item?id=515df4c84d85f53c2811
        else:
            am = AppMaterial(nxurl)
#             print(nxurl)
            am.handleM(title)
            material = am.material
            tdict['material'] = material

    clst.append(tdict)
#     except:
#         print('---'+str(i)+'----------')
        
    time.sleep(1)
    
#     print(tdict)
#     print('******************')
    
#     print(plist.content)
```
