```python
path = "../data/"
filename = 'result.json'

with open(path+filename,'r',encoding='utf-8') as fin:
    rawdata = json.load(fin)
fin.close()
```

```python
nds = []
nd = {}
```

```python
### 政策id是主码 自己定义 每次自增即可
## title 文件标题
## agent 发布机构
## dept_code   部门（编码值ID吗）
## level_code  级别 (编码值ID吗)              ## 通过数据的key = area属性判断 最后一个字 
# rdate  '发布日期',
# sdate      '开始时间',
# edate     '结束时间',
# tocde '适用对象',                            ## 除了标题中包括：人才认定、人才评定为个人外，其他的都设置为企业
# area "适用地区"
# POLICY_CODE          '文号',
#    KEYWORD         '关键字'              取空 不用插入
#    topic        '主题',
#    dire     '扶持方向',
#    CONTENT            '正文内容',
#    POLICY_STATUS       '状态',           取空 不用插入
#    ANNEX_URL   '附件',                   这个值是得内容是一个列表 插入数据库时注意 可能要把它转化为字符串
# POLICY_DECO        '政策分解',          取空 不用插入
#    VALID_FLAG  '有效标识',
#    CREATE_USER           '创建人',     ## 这个以下的四个值得赋值方式类似于上一次
#    CREATE_TIME           '创建时间',
#    UPDATE_USER           '最后更新人',
#    UPDATE_TIME           '最后更新时间',
for rd in rawdata:
    

    nd['title'] = rd['title']
    # nd['agent'] = rd['agent']


    dept1 = rd['agent']
#     print(dept1)
    if '其他' not in dept1:
        dept = dept1
    else:
        agent,agent_nto,agent_all = get_agent(rd)
        dept = get_dept(agent,agent_nto,agent_all)
#     print(dept)
    nd['agent'] = dept


    nd['dept_code'] = 'BMZ016'



    nd['level_code'] = get_level(rd['area'])
    # if rd['time'] == "":
    #     nd['rdate'] = '0000-00-00'
    # else:
    #     nd['rdate'] = rd['time']
    nd['rdate'],nd['sdate'],nd['edate'] = get_vdata(rd['time'],rd['timespan'])

    # nd['sdate'] = '2019-10-17'
    # nd['edate'] = '2019-10-17'
    nd['tcode'] = get_tcode(rd['title'])
    nd['area'] = rd['area']
    nd['policy_code'] = rd['artid']
    nd['topic'] = get_theme(rd['title']+rd['mainbody'])
    nd['dire'] = get_supportdirection(rd['title']+rd['mainbody'])
    nd['content'] = rd['mainbody']
    nd['annex_url'] = rd['appendix']
    nd['valid_flag'] = str(get_validflag(rd['timespan']))
    
    print(nd['agent'])
    nds.append(nd)
```

```python
# nds.append(nd)

with open(path+'test2.json','w+',encoding='utf-8') as fout:
    json.dump(nds,fout,ensure_ascii=False,indent=4)
```
